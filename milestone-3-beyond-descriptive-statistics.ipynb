```python
import pandas as pd

# Replace NaN values with 0 in numeric columns (optional)
numeric_columns = ['Age', 'Height', 'Weight','Year']
df_athletes_events[numeric_columns] = df_athletes_events[numeric_columns].fillna(0)

# Calculate and print interquartile range for each selected column
for selected_column in numeric_columns:
    # Remove NaN values from the selected column
    selected_column_data = df_athletes_events[selected_column].dropna()

    # Calculate the first quartile (Q1)
    q1 = selected_column_data.quantile(0.25)

    # Calculate the third quartile (Q3)
    q3 = selected_column_data.quantile(0.75)

    # Calculate the interquartile range (IQR)
    iqr = q3 - q1

    print(f"Interquartile Range of {selected_column}: {iqr:.2f}")
```

```python
import pandas as pd
from pandasql import sqldf
from scipy import stats

# Replace NaN values with 0 in numeric columns (optional)
numeric_columns = ['Age', 'Height', 'Weight','Year']
df_athletes_events[numeric_columns] = df_athletes_events[numeric_columns].fillna(0)

# Iterate through each selected column and calculate statistics
for selected_column in numeric_columns:
    # Calculate mean
    mean = df_athletes_events[selected_column].mean()

    # Calculate median
    median = df_athletes_events[selected_column].median()

    # Calculate mode
    mode = stats.mode(df_athletes_events[selected_column])[0][0]

    print(f"Statistics for {selected_column}:")
    print(f"Mean: {mean:.2f}")
    print(f"Median: {median:.2f}")
    print(f"Mode: {mode:.2f}")
    print("") 
```

```python
# Group data by 'NOC' (Country) and count the unique IDs (participants)
participation_count = df_athletes_events.groupby('NOC')['ID'].nunique()

# Group data by 'NOC' and 'Medal', then count the occurrences of each medal type
medal_counts = df_athletes_events.groupby(['NOC', 'Medal'])['ID'].count().unstack(fill_value=0)

# Calculate total medals won by each country
total_medals = medal_counts.sum(axis=1)

# Calculate the correlation between participation count and total medals won
correlation = participation_count.corr(total_medals)

# Print the correlation
print("Correlation between Participation Count and Total Medals Won:", correlation)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create a correlation heatmap using Seaborn
plt.figure(figsize=(10, 6))
sns.heatmap(pd.DataFrame({'Participation Count': participation_count, 'Total Medals': total_medals}).corr(),
            annot=True, cmap='coolwarm', linewidths=.5)
plt.title(f'Correlation Map: Participation Count vs. Total Medals (Correlation: {correlation:.2f})')
plt.show()
```

```python
# Filter data for Gold medals
gold_medals = df_athletes_events[df_athletes_events['Medal'] == 'Gold']

# Group data by 'NOC' (Country) and count the unique IDs (participants)
participation_count = df_athletes_events.groupby('NOC')['ID'].nunique()

# Group data by 'NOC' and count the occurrences of Gold medals
gold_medals_count = gold_medals.groupby('NOC')['ID'].count()

# Merge participation and gold medals data
correlation_data = pd.DataFrame({'Participation Count': participation_count, 'Gold Medals Count': gold_medals_count})
correlation_data = correlation_data.dropna()

# Calculate the correlation between participation count and gold medals count
correlation = correlation_data['Participation Count'].corr(correlation_data['Gold Medals Count'])

# Create a scatter plot using Seaborn
plt.figure(figsize=(10, 6))
sns.scatterplot(data=correlation_data, x='Participation Count', y='Gold Medals Count')
plt.title(f'Correlation: Participation Count vs. Gold Medals Count (Correlation: {correlation:.2f})')
plt.xlabel('Participation Count')
plt.ylabel('Gold Medals Count')
plt.show()
```

```python

# Group data by 'NOC' (Country) and count the unique IDs (participants)
participation_count = df_athletes_events.groupby('NOC')['ID'].nunique()

# Group data by 'NOC' and count the occurrences of each medal type
medal_counts = df_athletes_events.groupby(['NOC', 'Medal'])['ID'].count().unstack(fill_value=0)

# Calculate total medals won by each country
total_medals = medal_counts.sum(axis=1)

# Merge participation and total medals data
correlation_data = pd.DataFrame({'Participation Count': participation_count, 'Total Medals': total_medals})
correlation_data = correlation_data.dropna()

# Calculate the correlations
correlation_gold = correlation_data['Participation Count'].corr(correlation_data['Total Medals'])
correlation_silver = correlation_data['Participation Count'].corr(medal_counts['Silver'])
correlation_bronze = correlation_data['Participation Count'].corr(medal_counts['Bronze'])

# Create a scatter plot using Seaborn with different markers for each medal type
plt.figure(figsize=(10, 6))
sns.scatterplot(data=correlation_data, x='Participation Count', y='Total Medals', label=f'Gold (Correlation: {correlation_gold:.2f})', marker='o')
sns.scatterplot(data=correlation_data, x='Participation Count', y=medal_counts['Silver'], label=f'Silver (Correlation: {correlation_silver:.2f})', marker='s')
sns.scatterplot(data=correlation_data, x='Participation Count', y=medal_counts['Bronze'], label=f'Bronze (Correlation: {correlation_bronze:.2f})', marker='^')
plt.title('Correlation: Participation Count vs. Medals Count')
plt.xlabel('Participation Count')
plt.ylabel('Medals Count')
plt.legend()
plt.show()
```

```python
# Group data by year and country and count the number of participants
participants_by_year_country = df_athletes_events.groupby(['Year', 'NOC'])['ID'].nunique().unstack()

# Create a stacked area plot
plt.figure(figsize=(12, 8))
participants_by_year_country.plot(kind='area', colormap='tab20')
plt.title('Participation by Country in the Olympics Over Time')
plt.xlabel('Year')
plt.ylabel('Number of Participants')
plt.legend(title='Country', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
# Create a pivot table to count the number of participants by year and gender
participants_by_gender = df_athletes_events.pivot_table(index='Year', columns='Sex', values='ID', aggfunc='nunique')

# Create a line plot to visualize participation of men and women over the years
plt.figure(figsize=(10, 6))
participants_by_gender.plot(kind='line')
plt.title('Participation of Men and Women in Olympics Over the Years')
plt.xlabel('Year')
plt.ylabel('Number of Participants')
plt.legend(title='Gender')
plt.show()
```

```python
# Group data by year and count the number of women participants and medals won
women_participation_and_medals = df_athletes_events[df_athletes_events['Sex'] == 'F'].groupby('Year').agg(
    {'ID': 'nunique', 'Medal': 'count'}).rename(columns={'ID': 'Women Participants', 'Medal': 'Medals Won'})

# Calculate the correlation coefficient
correlation_coefficient = women_participation_and_medals['Women Participants'].corr(women_participation_and_medals['Medals Won'])

print(f"Correlation between Women Participation and Medals Won: {correlation_coefficient:.2f}")

# Create a scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(women_participation_and_medals['Women Participants'], women_participation_and_medals['Medals Won'],
            color='b', alpha=0.7)
plt.title('Correlation: Women Participation vs. Medals Won')
plt.xlabel('Number of Women Participants')
plt.ylabel('Number of Medals Won')
plt.grid(True)
plt.show()
```

```python
# Group data by year and count the number of men participants and medals won
men_participation_and_medals = df_athletes_events[df_athletes_events['Sex'] == 'M'].groupby('Year').agg(
    {'ID': 'nunique', 'Medal': 'count'}).rename(columns={'ID': 'Men Participants', 'Medal': 'Medals Won'})

# Calculate the correlation coefficient
correlation_coefficient = men_participation_and_medals['Men Participants'].corr(men_participation_and_medals['Medals Won'])

print(f"Correlation between Men Participation and Medals Won: {correlation_coefficient:.2f}")


# Create a scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(men_participation_and_medals['Men Participants'], men_participation_and_medals['Medals Won'],
            color='b', alpha=0.7)
plt.title('Correlation: Men Participation vs. Medals Won')
plt.xlabel('Number of Men Participants')
plt.ylabel('Number of Medals Won')
plt.grid(True)
plt.show()
```
